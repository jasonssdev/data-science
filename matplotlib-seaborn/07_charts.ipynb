```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.random.randint(1,50,100)
```

# Histograma

```python
plt.hist(data, bins=10, histtype='bar')
plt.show()
```

```python
plt.hist(data, bins=10, histtype='step')
plt.show()
```

# Boxplot

```python
plt.boxplot(data)
plt.show()
```

```python
# adding a outlier data
data = np.append(data, 200)
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=True)
plt.show()
```

# Scatter

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * x) ** 2
colors =  y
```

```python
plt.scatter(x,y, s=area, c=colors, marker='o', alpha=0.5)
plt.show()
```
