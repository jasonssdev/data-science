```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, "r")
axes.plot(y,x, "b--")
fig.show()
```

```python
fig = plt.figure(facecolor='gray')
axes1 = fig.add_axes([0.1,0.1,0.8,0.9])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes1.plot(x,y, "r")
axes2.plot(y,x, "b--")
axes2.set_facecolor("yellow")
fig.show()
```
