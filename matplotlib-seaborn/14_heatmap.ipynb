```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips_corr = tips.corr(numeric_only = True)
```

```python
sns.heatmap(tips_corr, annot=True, cmap='coolwarm', linewidths=5, linecolor='grey', vmin=-1, vmax=1)
plt.show()
```

```python
sns.heatmap(tips_corr, annot=True, cmap='coolwarm', linewidths=5, linecolor='grey', vmin=0.5, vmax=1, cbar=True)
plt.show()
```
