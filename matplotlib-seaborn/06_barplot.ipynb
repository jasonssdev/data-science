```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
country_list = ['BRAZIL', 'JAPAN', 'MEXICO', 'COLOMBIA', 'USA']
population_list = [1000, 800, 1000, 500, 800]
```

```python
plt.bar(country_list, population_list, width=0.5, color=['red', 'blue', 'green', 'purple', "grey"])
plt.xticks(np.arange(5),('Brasil', 'Japon', 'Mexico', 'Colombia', 'Estados Unidos'), rotation=45)
plt.show()
```

```python
plt.barh(country_list, population_list ,height=0.5, color=['red', 'blue', 'green', 'purple', "grey"], label=country_list)
plt.title('Country Population chart')
plt.xlabel(xlabel='population')
plt.ylabel(ylabel='countries')
plt.legend(loc='center', bbox_to_anchor=(1.15, 0.5))
plt.yticks(np.arange(5),('Brasil', 'Japon', 'Mexico', 'Colombia', 'Estados Unidos'), rotation=0)
plt.show()
```
