```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
```

```python
print(plt.style.available)
```

```python
plt.style.use("dark_background")
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,x+1,'r--')
ax.plot(x,x+2,'g-')
ax.plot(x,x+3,'y:')
ax.plot(x,x+4,'b.')


```

## Pyplot

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,x+1,color = 'green', alpha=0.5, linewidth = 5, linestyle='-', marker='o', markerfacecolor = 'red', markersize=10)
ax.plot(x,x+2,color = 'blue', alpha=0.8, linewidth = 5, linestyle='--')
ax.plot(x,x+3,color = 'red', alpha=0.5, linewidth = 8, linestyle='dashed')
ax.plot(x,x+4,color = '#66BADD', alpha=0.5, linewidth = 8, linestyle=':')
```
