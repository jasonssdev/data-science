```python
import numpy as np
import matplotlib.pyplot as plt 
```

# Function linear algebraic

```python
N = 100
m = 10
b = 5

def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)

y = f(x)
```

```python
y.shape
```

```python
fig , ax = plt.subplots()
ax.plot(x,y)
ax.grid()
```

# function polynomial algebraic

```python
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y = f(x)
plt.plot(x,y)
plt.grid()
plt.show()
```
