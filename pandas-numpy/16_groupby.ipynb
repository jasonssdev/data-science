```python
import pandas as pd
import numpy as np
```

```python
df_books = pd.read_csv("./data/bestsellers with categories.csv")
df_books.head()
```

```python
# pivot table
df_books.groupby("Author").count()
```

```python
# firter bygroup by william author with loc function
df_books.groupby("Author").count().loc["William Davis"]
```

```python
# to reset index
df_books.groupby("Author").count().reset_index()
```

```python
df_books.groupby("Author").agg(["min", "max"])
```

```python
df_books.groupby("Author").agg({"Reviews":["min", "max"], "Price":"sum"})
```

```python
df_books.groupby(["Author", "Year"]).agg({"Reviews":["min", "max"], "Price":"sum"})
```
