# Merge

```python
df_left = pd.DataFrame({"key":["k0","k1","k2","k3"],
"A":["A0","A1","A2","A3"],
"B":["B0","B1","B2","B3"]})
```

```python
df_right = pd.DataFrame({"key":["k0","k1","k2","k3"],
"C":["C0","C1","C2","C3"],
"D":["D0","D1","D2","D3"]})
```

```python
df_left
```

```python
df_right
```

```python
df_left.merge(df_right, on="key")
```

```python
df_left = pd.DataFrame({"key":["k0","k1","k2","k3"],
"A":["A0","A1","A2","A3"],
"B":["B0","B1","B2","B3"]})

df_right = pd.DataFrame({"key_2":["k0","k1","k2","k3"],
"C":["C0","C1","C2","C3"],
"D":["D0","D1","D2","D3"]})
```

```python
df_left.merge(df_right, left_on="key", right_on="key_2")
```

```python
df_left = pd.DataFrame({"key":["k0","k1","k2","k3"],
"A":["A0","A1","A2","A3"],
"B":["B0","B1","B2","B3"]})

df_right = pd.DataFrame({"key_2":["k0","k1","k2",np.nan],
"C":["C0","C1","C2","C3"],
"D":["D0","D1","D2","D3"]})
```

```python
# merge did not find key3
df_left.merge(df_right, left_on="key", right_on="key_2")
```

```python
# to solve use how parameters
df_left.merge(df_right, left_on="key", right_on="key_2", how="left")
```

```python
# to solve use how parameters
df_left.merge(df_right, left_on="key", right_on="key_2", how="right")
```

```python
# to solve use how parameters
df_left.merge(df_right, left_on="key", right_on="key_2", how="inner")
```

```python
# to solve use how parameters
df_left.merge(df_right, left_on="key", right_on="key_2", how="outer")
```

# Join - Index match

```python
df_left = pd.DataFrame({
"A":["A0","A1","A2"],
"B":["B0","B1","B2"]}, index=["k0","k1","k2"])

df_right = pd.DataFrame({
"C":["C0","C1","C2"],
"D":["D0","D1","D2"]}, index=["k0","k2","k3"])
```

```python
df_left
```

```python
df_right
```

```python
df_left.join(df_right)
```

```python
df_left.join(df_right, how="left")
```

```python
df_left.join(df_right, how="right")
```

```python
df_left.join(df_right, how="inner")
```

```python
df_left.join(df_right, how="outer")
```
